# Análise exploratória sobre o crescimento empresarial no Brasil: Fatores econômicos, financeiros e desenvolvimento sustentável afetam o empreendedorismo por oportunidade?

Ariano Bangemann

Há um relativo consenso na literatura de que os empreendedores são motivados por dois grandes grupos de fatores ao empreender. Aqueles que por algum motivo encontram-se sem soluções e são empurrados ao empreendedorismo, seja pela falta de opções, pela informalidade de suas atividades ou pela insatisfação com o trabalho atual e que são classificados como empreendedores por necessidade. E um segundo grupo na qual vê no empreendedorismo como algo favorável a uma realização, que são atraídos e impulsionados por mais opções, que é o empreendedorismo por oportunidade (Bosma e Harding, 2007; Bygrave et. al. 2003; Maritz, 2004).

Em seu artigo, Cervelló-Royo et.al. (2020), buscam identificar fatores que podem estar relacionados com o empreendedorismo por oportunidade em determinados países. Os autores argumentam que a combinação de fatores como o desenvolvimento sustentável, aspectos econômicos e financeiros podem potencializar a atividade empreendedora. Para tanto, analisam dados de 57 países referentes ao ano de 2017 extraídos do GEM - Global Entrepreneurship Monitor, do Country Risk Score, do World Bank Database e do Sustainable Development Goals Index utilizando o método fsQCA.

Como resultado, Cervelló-Royo et.al. (2020), concluem que três combinações são adequadas para relacionar a atividade empreendedora com as variáveis preditoras. Duas delas, PIB per capita e risco país (CRS) estão presentes nas três combinações. Inferem ainda, que para que uma alta oportunidade ocorra, é necessário acesso a uma boa situação econômica e financeira, com acesso a mercados financeiros internacionais. Também é necessário que um país crie riqueza e que sua educação seja de alta qualidade, além de ter uma indústria desenvolvida, assim como inovação e infraestrutura. Com estes preceitos, sugerem que se invista no empreendedorismo por oportunidade.

O empreendedorismo por oportunidade tem sido objeto de uma gama de estudos, uma vez que pesquisadores buscam compreender quais os antecedentes que promoveram a escolha e qual sua relação com as possíveis variáveis motivadoras. Os empreendedores por oportunidade estão em busca de desafios e são movidos por interesses particulares que surgem a partir de determinados gatilhos. Entretanto, fatores ambientais internos e externos podem provocar – ou não – o disparo desta propensão. A grande dúvida que cerca os estudos é descobrir quais são exatamente estes gatilhos, quais os fatores ambientais internos e externos que atuam nestas decisões e até que ponto os interesses particulares podem ser afetados por estas variáveis, já que a literatura não tem um senso comum se oportunidades são criadas ou descobertas (Ardichvili et. al., 2003; Short et. al., 2010).

Alguns autores, porém, discordam da separação em dois segmentos motivacionais para o empreendedorismo. Acreditam que até mesmo um empreendedor movido pela necessidade, pode vislumbrar uma oportunidade de negócio, já que os trabalhadores informais estão incluídos no grupo dos empreendedores por necessidade. Esta informalidade pode surgir por fatores diversos, sejam eles de qualidade econômica, financeira, estrutural ou legal. Deste modo, podemos ainda inferir que aqueles fatores que afetam o empreendedorismo por oportunidade, podem ser os mesmos ou parte dos que afetam o empreendedorismo por necessidade (Williams, 2008).

Neste caso, podemos considerar que se temos uma divisão em dois grandes segmentos motivacionais para o empreendedorismo, necessidade e oportunidade, fatores internos e externos podem afetar diferentemente a decisão e a forma de empreender. Por outro lado, se não podemos separar as motivações de necessidade e oportunidade por estarem relacionadas, precisamos compreender uma gama maior de fatores que podem afetar o empreendedorismo, muito mais que fatores ambientais externos.

Afsar et. al. (2021) acreditam que o empreendedorismo por oportunidade está relacionado a outros fatores diferentes que não podem ser previstos ou classificados e que se comportam melhor quando há a possibilidade de sistemas de recomendação atuarem. Mesmo os métodos tradicionais utilizam conceitos relacionados a filtragem colaborativa ou a filtragem de conteúdo. Deste modo, os autores sugerem que algoritmos de reinforcement learning baseados no processo de decisão de Markov (MDP) podem se comportar melhor na assertividade do empreendedorismo por oportunidade.

Assim, neste artigo, este autor questiona as conclusões que os autores Cervelló-Royo et.al. (2020) aduzem. Primeiro porque não há consenso quanto as relações entre necessidade e oportunidade. O empreendedorismo gerado pela informalidade pode estar amparado por uma oportunidade que está subjugada por diversas motivações, dentre elas as de fator econômico e financeiro, que são objeto do estudo daqueles autores.

Segundo, Cervelló-Royo et. al. reconhecem a necessidade de aprofundamento dos estudos futuros. O estudo apresenta uma gama ínfima de países em sua amostra, como os próprios autores reconhecem, e que não formam uma base sustentável para formar preditores do empreendedorismo. Acredito que mesmo dentro dos países propostos, deve-se ter diferentes áreas com condições adversas econômica e financeiramente, e com práticas voltadas a sustentabilidade divergentes. Países são divididos em estados, regiões e municípios, muitos com condições e estruturas semelhantes de afinidade, mesmo que distantes e até mesmo bairros e localidades vizinhas com condições extremamente diferentes. Podemos citar Brasil e China como países de imensidão territorial e com peculiaridades próprias de cada região.

Terceiro, a gama de variáveis propostas no estudo representam apenas uma pequena parcela das possibilidades que podem influenciar potencialmente o empreendedorismo. Este autor acredita que o empreendedorismo e as oportunidades podem ser criados ou descobertos motivados por fatores externos como é o caso do estudo de Cervelló-Royo et. al., porém fatores intrínsecos ao empreendedor, ao contexto, as descobertas, ao momento e a uma gama ímpar de condições afetam este estado. Aquilo que serviu em dado momento pode ser reprocessado e criadas novas condições que vão alterar as próximas oportunidades. Somente o aprendizado de reforço será capaz de atuar de forma autônoma em cada momento e para cada empreendedor.

### 10.2 Mínimos Quadrados Ordinários (RSS)

```python
# Regression coefficients (Ordinary Least Squares)
regr = skl_lm.LinearRegression()

X = scale(df.plan, with_mean=True, with_std=False).reshape(-1,1)
y = df.vd_emp

regr.fit(X,y)
print(regr.intercept_)
print(regr.coef_)
```

```python
# Create grid coordinates for plotting
B0 = np.linspace(regr.intercept_-2, regr.intercept_+2, 50)
B1 = np.linspace(regr.coef_-0.02, regr.coef_+0.02, 50)
xx, yy = np.meshgrid(B0, B1, indexing='xy')
Z = np.zeros((B0.size,B1.size))

# Calculate Z-values (RSS) based on grid of coefficients
for (i,j),v in np.ndenumerate(Z):
    Z[i,j] =((y - (xx[i,j]+X.ravel()*yy[i,j]))**2).sum()/1000

# Minimized RSS
min_RSS = r'$\beta_0$, $\beta_1$ for minimized RSS'
min_rss = np.sum((regr.intercept_+regr.coef_*X - y.values.reshape(-1,1))**2)/1000
min_rss
```

```python
fig = plt.figure(figsize=(15,6))
fig.suptitle('RSS - Regression coefficients', fontsize=20)

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

# Left plot
CS = ax1.contour(xx, yy, Z, cmap=plt.cm.Set1, levels=[1.15, 1.2, 1.3, 1.4, 2])
ax1.scatter(regr.intercept_, regr.coef_[0], c='r', label=min_RSS)
ax1.clabel(CS, inline=True, fontsize=10, fmt='%1.1f')

# Right plot
ax2.plot_surface(xx, yy, Z, rstride=3, cstride=3, alpha=0.3)
ax2.contour(xx, yy, Z, zdir='z', offset=Z.min(), cmap=plt.cm.Set1,
            alpha=0.4, levels=[1.15, 1.2, 1.3, 1.4, 2])
ax2.scatter3D(regr.intercept_, regr.coef_[0], min_rss, c='r', label=min_RSS)
ax2.set_zlabel('RSS')
ax2.set_zlim(Z.min(),Z.max())
ax2.set_ylim(0.17,0.22)

# settings common to both plots
for ax in fig.axes:
    ax.set_xlabel(r'$\beta_0$', fontsize=17)
    ax.set_ylabel(r'$\beta_1$', fontsize=17)
    ax.set_yticks([0.18,0.19,0.20,0.21])
    ax.legend()
```

Este estudo buscou demonstrar que o empreendedorismo por oportunidade não se limita a determinadas variáveis e não pode ser motivado apenas por questões econômicas, financeiras ou de desenvolvimento sustentável como proposto por Cervelló-Royo et. al. (2020).  Apesar dos autores demonstrarem através da metodologia fsQCA que o PIB per capita e o risco país estão altamente ligados ao empreendedorismo por oportunidade, a amostra do estudo foi limitada e não pode representar um modelo abrangente.

Utilizei um algoritmo "ensemble" combinando predições para gerar correlações entre uma variável dependente (vd_emp) – crescimento anual no número de empresas em determinado município – e outras 8 variáveis independentes. Dentro destas variáveis trabalhamos com possíveis motivações financeiras, econômicas e de desenvolvimento sustentável, assim como Cervelló-Royo et. al. (2020) – PIB per capita, crescimento médio do PIB e qualidade da educação – e que poderiam impactar positivamente o empreendedorismo por oportunidade. Além destas, buscamos agregar outras consideradas fomentadoras do empreendedorismo como meio-ambiente e desenvolvimento sustentável, tecnologia, gestão, articulação interinstitucional e instrumentos do planejamento.

Mesmo que correlações entre a variável dependente e as variáveis planejamento, crescimento médio do PIB e desenvolvimento sustentável possam ter se apresentado fortes, não foram determinantes para se afirmar uma causalidade entre elas. Observamos um baixo impacto no empreendedorismo quando utilizamos uma amostra maior (5.569 municípios brasileiros) e que não representam congruências para todos os objetos de estudo. Observamos inclusive que quando há uma combinação de variáveis, que é o caso de desenvolvimento sustentável, planejamento e crescimento médio do PIB, mesmo assim não há uma forte correlação que sustente a determinação das motivações para o empreendedorismo.

Deste modo, não se pode afirmar que há forte causalidade entre fatores motivacionais para o empreendedorismo por oportunidade relacionados a fatores econômicos, financeiros e de desenvolvimento sustentável como propõe Cervelló-Royo et. al., (2020). O empreendedorismo recebe influência destes fatores, como comprovado em nosso estudo, porém não são determinantes para o sucesso. Acreditamos que fatores intrínsecos a cada empreendedor sejam mais fortes e determinantes do que propriamente os fatores ambientais externos. Um empreendedor pode ter mais sucesso mesmo estando em localidades de baixo impacto em relação a outro que esteja em condições externas mais favoráveis. Somente algoritmos que possam processar e se adaptar rapidamente a estes fatores podem prever com maior sucesso a atividade empreendedora.
